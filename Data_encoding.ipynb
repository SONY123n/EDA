```python
df=pd.DataFrame({'time':['lunch','breakfast','dinner','breakfast','lunch','dinner','lunch','breakfast','dinner'],
             'total_bill':[120,130,40,150,160,130,150,100,160]})
```

```python
df
```

```python
mean_price=df.groupby('time')['total_bill'].mean().to_dict()
```

```python
mean_price
```

```python
df['time_target_encoder']=df['time'].map(mean_price)
```

```python
df
```
