```python
import numpy as np 
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
np.random.seed(1) #for reproducibility
```

```python
no_samples=900
class_0_ratio = 0.9
no_class_0=int(no_samples + class_0_ratio)
no_class_1 =100
```

```python
class_0_ratio
```

```python
no_class_0
```

```python
no_class_1
```

```python
np.random.normal(0,1,no_class_0)
```

```python
len(np.random.normal(0,1,no_class_0))
```

```python
[0]*10
```

```python
class_0={'feature1': np.random.normal(0,1,no_class_0),
 'feature2': np.random.normal(0,1,no_class_0),
 'target':[0]*no_class_0}
```

```python
class_0
```

```python
class_0=pd.DataFrame(class_0)
```

```python
class_0
```

```python
class_1=pd.DataFrame({'feature1': np.random.normal(3,1,no_class_1),
 'feature2': np.random.normal(3,1,no_class_1),
 'target':[1]*no_class_1})
```

```python
class_1
```

```python
df=pd.concat([class_0,class_1]).reset_index(drop=True)
```

```python
df
```

```python
df.target.value_counts()
```

```python
df.target.value_counts(normalize=True)#means this is a classic case of class imbalnce
```

```python
df_minority =df[df.target == 1]
df_majority=df[df.target == 0]
```

```python
df_minority
```

```python
df_majority
```

```python
#upsampling or over sampling >> u increase the minority data to majority 
from sklearn.utils import resample
```

```python
df_minority_upsample=resample(df_minority,replace=True,n_samples=len(df_majority),random_state=1)
```

```python
df_minority_upsample
```

```python
df_minority_upsample.shape
```

```python
df_minority_upsample.duplicated().sum()
```

```python
df_upsample=pd.concat([df_majority,df_minority_upsample])
```

```python
df_upsample
```

```python
df_upsample.shape
```

```python
df_upsample.target.value_counts()
```

```python
#downsampling
```

```python
df_minority
```

```python
df_majority
```

```python
df_majority_downsample=resample(df_majority,replace=False,n_samples=len(df_minority),random_state=1)
```

```python
df_majority_downsample
```

```python
df_downsample=pd.concat([df_minority,df_majority_downsample])
```

```python
df_downsample
```

```python
#SMOTE >> synthetic minority oversampling technique
```

```python
from sklearn.datasets import make_classification 
```

```python
x,y=make_classification(n_samples=1000,n_redundant=0,n_features=2,n_clusters_per_class=1,weights=[0.90],random_state=1)
```

```python
x
```

```python
y
```

```python
len(y[y==0])
```

```python
len(y[y==1])
```

```python
df1=pd.DataFrame(x,columns=['f1','f2'])
df2=pd.DataFrame(y,columns=['target'])
df_final=pd.concat([df1,df2],axis=1)
```

```python
df_final
```

```python
df_final.target.value_counts()
```

```python
df_final
```

```python
plt.scatter(df_final['f1'],df_final['f2'],c=df_final['target'])
```
